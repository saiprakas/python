```python
class A:

    def __init__(self,a,b):
        self.a = a
        self.b = b
        print(a+b)
a1 = A(2,3)
```

```python
class B:
    def __init__(self,a):
        fact=1
        for i in range(1,a+1):
            fact=fact*i
        print(fact)
a = B(4)
```

```python
class A:
    def perfect(self,a):
        sum=0
        for i in range (1,a):
            if(a%i==0):
                sum+=i
        if sum==a:
            print("perfect")
        else:
            print("not perfect")
    def prime(self,c):
        if(c==2 or c==5 or c==7):
            print("prime number")
        elif(c%1==0 and c%c==0 and c%2!=0 and c%5!=0 and c%7!=0):
            print("prime number")
        else:
            print("not a prime number")
a=A()
a.perfect(6)
a.prime(13)
```
